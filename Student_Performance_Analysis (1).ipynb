# Student Performance Analysis

### Univariate analysis

##Average Final Grade (G3)
- Calculating the mean of the final grade column (G3)

```python
round(df["G3"].mean(),2)
```

###Insight:

- The mean of the final grade column (G3) is 10.42

- The average final grade represents the overall academic performance of students. It provides a benchmark to evaluate whether student performance is generally low, moderate, or high.

#Filtering

##Students Scoring Above 15
- Counting students whose final grade (G3) is greater than 15.

```python
df[df["G3"] > 15].shape[0]
```

###Explanation

- operator > Filters students with grades above 15.

- .shape Counts such students.

###Insight

- There are 40 students whose final grade (G3) is greater than 15.

- Students scoring above 15 are considered high achievers. This metric helps identify the proportion of students performing exceptionally well and gives an idea of academic excellence

# Sorting

## Gender-wise Performance

- Calculating average final grades grouped by gender

```python
df.groupby("sex")["G3"].mean()
```

###Explanation

- .groupby Groups students by gender.

- .mean Calculates average grade for each group.


###Insight:

- The results show that female students perform slightly better on average than male students.It highlights performance differences and can help educators understand demographic trends in academic outcomes.

#Study Time vs Performance

###Multivariate analysis

```python
df["studytime"].corr(df["G3"])
```

###Explanation

- .corr Computes correlation between study time and grades.

### Insight

- A positive correlation indicates that students who spend more time studying tend to score higher grades. Although correlation does not imply causation, it strongly suggests that consistent study habits positively impact academic performance.

#Visualizations

#Histogram of Final Grades
- Plotting a histogram for final grades.

```python
import matplotlib.pyplot as plt

plt.hist(df["G3"])
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades")
plt.show()
```

Insight

Most students score between 10 and 15, showing a concentration in the mid-grade range.Very few students fall at the extreme low or high ends, indicating a normal distribution of academic performance

#Scatter Plot: Study Time vs Grades

- Createing a scatter plot between study time and final grades

```python
plt.scatter(df["studytime"], df["G3"])
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Study Time vs Performance")
plt.show()
```

###Insight

- The upward trend in the plot confirms that higher positive relationship between study time and academic performance. generally corresponds to better grades. However, variations indicate that other factors also influence student academic performance.



# Bar Chart: Gender vs Average Grade

- Plotted a bar chart comparing average grades of male and female students.

```python
df.groupby("sex")["G3"].mean().plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.title("Average Grade by Gender")
plt.show()
```

###Insight:

- Female students show a higher average final grade compared to male students, reinforcing the numerical analysis and making the comparison visually clear and easy to interpret
